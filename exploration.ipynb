```python
names = df_otu_names.columns.tolist()
name_to_drop = 'Order'

flag=True
cols_to_drop = []
for name in names:
    if name != name_to_drop and flag==True:
        continue
    flag=False
    cols_to_drop.append(name)
    
# inicjalizacja mapy OTU
otu_map = {}
for otu in df_otu_names['Unnamed: 0'].values:
    otu_map[otu] = otu

# inicjalizacja mapy kolumn do OTU
col_map = {}

# usun kolumny ktore nie sa potrzebne
df = df_otu_names.drop(columns=cols_to_drop)


# iterujemy po wierszach i tworzymy wpisy w mapie column do OTU
for row in df.itertuples(name=None, index=False):
    entry = ''
    for name in row[1:]:
        entry = entry + str(name) + ';'
        
    # jesli juz jest taki wpis to w mapie OTU na OTU dodajemy OTU z rowa -> OTU z wpisu z mapy kolumn na OTU
    if entry in col_map:
        otu_map[row[0]] = col_map[entry]
    # jezeli jeszcze nie bylo takiego wpisu to dodajemy do mapy kolumn na OTU
    else:
        col_map[entry] = row[0]
```

```python
missin_no = 0 
df = df_otu_table.copy()
# usuwamy wszystkie kolumny ktore maja same zera
df = df.loc[:, (df != 0).any(axis=0)]
# iterujemy po mapie OTU
for k,v in otu_map.items():
    # jesli nie ma podrzednego otu w tym df to pomijamy
    if k not in df.columns:
        continue
    # jesli nie ma nadrzednego to tworzymy nowa kolumne z zerami dla niego
    if v not in df.columns:
        df[v] = 0
        missin_no += 1
    # jesli k i v takie same to pomijamy zeby nie dodawac do siebie samego
    if k == v:
        continue
    # jesli sa rozne to mamy poprawne mapowanie i dodajemy wartosci z k do v, po czym usuwamy k z df
    else:
        df[v] = df[v] + df[k]
        df = df.drop(columns=[k])
        
# jeszcze raz usuwany kolumny z samymi zerami jakby sie okazalo ze jakies stworzylismy po drodze
df = df.loc[:, (df != 0).any(axis=0)]

print(f'ilosc brakujacych OTU: {missin_no}')
df 
```
